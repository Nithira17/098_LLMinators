```python
from langchain_google_genai import ChatGoogleGenerativeAI,GoogleGenerativeAIEmbeddings
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')


llm=ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    temperature=0,
    api_key=GOOGLE_API_KEY
)

embedding_model= GoogleGenerativeAIEmbeddings(
    model="models/embedding-001"
)
```

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings

# Used HuggingFace embedding model as the embedding model
embedding_model_2 = HuggingFaceInstructEmbeddings(model_name="hkunlp/instructor-xl")
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

# Initialize DuckDuckGo search tool
search_tool = DuckDuckGoSearchRun()

search_tool
```

```python
from langchain.agents import load_tools
```

```python
from langchain_community.document_loaders import PyPDFLoader

# Load the PDF document
def load_doc(pdf_path):
  loader = PyPDFLoader(pdf_path)
  docs = loader.load()
  return docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Splitting the pdf
def text_splitter(docs):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=50)
  splits = text_splitter.split_documents(docs)
  return splits
```

```python
from langchain_chroma import Chroma

# store the chunks in vectorstore
def vector_store(splits):
  vectorstore = Chroma.from_documents(documents=splits, embedding=embedding_model_2)
  return vectorstore
```

```python
# Retriever function
def retriever(vectorstore):
  retriever = vectorstore.as_retriever()
  return retriever
```

```python
from langchain.chains import RetrievalQA

# RAG chain
def rag_chain(retriever):
  rag_chain = RetrievalQA.from_chain_type(
                                      llm=llm,
                                      chain_type="stuff",
                                      retriever=retriever)
  return rag_chain
```

```python
from langchain.agents import Tool

# Creating RAG tool
def rag_tool(rag_chain, RAG_tool, description):
  rag_tool = Tool(RAG_tool,
                func = rag_chain.run,
                description = description)

  return rag_tool
```

```python
# Ranil manifesto
r_pdf = load_doc('/content/manifesto-ranil.pdf')
r_text_split = text_splitter(r_pdf)
r_vectorstore = vector_store(r_text_split)
r_retriever = retriever(r_vectorstore)
r_ragchain = rag_chain(r_retriever)
RW_tool = rag_tool(r_ragchain, "Ranil Wickramasinghe Manifesto Tool", "Use when the user asks about Ranil Wickremesinghe's manifesto and policies. Also use when user asks to compare his policies with another candidate")

# Anura manifesto
a_pdf = load_doc('/content/manifesto-anura.pdf')
a_text_split = text_splitter(a_pdf)
a_vectorstore = vector_store(a_text_split)
a_retriever = retriever(a_vectorstore)
a_ragchain = rag_chain(a_retriever)
AKD_tool = rag_tool(a_ragchain,  "Anura Kumara Dissanayake Manifesto Tool", "Use when the user asks about Anura Kumara Dissanayake's or NPP's manifesto and policies. Also use when user asks to compare his policies with another candidate")

# Sajith manifesto
s_pdf = load_doc('/content/manifesto-sajith.pdf')
s_text_split = text_splitter(s_pdf)
s_vectorstore = vector_store(s_text_split)
s_retriever = retriever(s_vectorstore)
s_ragchain = rag_chain(s_retriever)
SP_tool = rag_tool(s_ragchain, "Sajith Premadasa Manifesto Tool", "Use when the user asks about Sajith Premadasa's or SJB's manifesto and policies. Also use when user asks to compare his policies with another candidate")

# Polls/predictions data
p_pdf = load_doc('/content/Polls.pdf')
p_text_split = text_splitter(p_pdf)
p_vectorstore = vector_store(p_text_split)
p_retriever = retriever(p_vectorstore)
p_ragchain = rag_chain(p_retriever)
POLL_tool = rag_tool(p_ragchain,  "Pre-election polls and prediction Tool", "Use when the user asks about polls and surveys done and their predictions")

# General elections data
e_pdf = load_doc('/content/general election questions.pdf')
e_text_split = text_splitter(e_pdf)
e_vectorstore = vector_store(e_text_split)
e_retriever = retriever(e_vectorstore)
e_ragchain = rag_chain(e_retriever)
ELECTION_tool = rag_tool(e_ragchain, "General election data Tool", "Use when the user asks questions about election")
```

```python
# Set prompt template
from langchain_core.prompts import PromptTemplate

template= '''
          Answer the following questions as best you can. You have access to the following tools:

          {tools}

          Use the following format:

          Question: the input question you must answer
          Thought: you should always think about what to do
          Action: the action to take, should be one of [{tool_names}]. Always look first in Vector Store
          Action Input: the input to the action
          Observation: the result of the action
          ... (this Thought/Action/Action Input/Observation can repeat 2 times)
          Thought: I now know the final answer
          Final Answer: the final answer to the original input question

          Begin!

          Question: {input}
          Thought:{agent_scratchpad}
          '''

prompt = PromptTemplate.from_template(template)
```

```python
from langchain.agents import create_react_agent

# List of tools for the agent
tools = [search_tool, RW_tool, AKD_tool, SP_tool, POLL_tool, ELECTION_tool]

# Create the agent using the LLM and the prompt template
agent = create_react_agent(tools = tools, llm = llm, prompt=prompt)
```

```python
from langchain.agents import AgentExecutor

# Initialize the agent executor with the created agent and tools
agent_executor = AgentExecutor(tools=tools, agent=agent, handle_parsing_errors=True, verbose=False)
```

```python
response = agent_executor.invoke({"input":"What is the Poll prediction in month august"})

print(response['output'])
```
